```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://movie.naver.com/movie/running/premovie.nhn"
page = urlopen(url)
soup = BeautifulSoup(page, 'lxml')
```

```python
soup.find("ul", class_="lst_detail_t1").find("dt",class_="tit").a.text
```

```python
soup.find_all("ul", class_="lst_detail_t1")
```

```python
infom= soup.find_all("ul", class_="lst_detail_t1")
infom[0].find_all("dt",class_="tit")
```

```python
soup.find("div", class_="article").find_all("dt",class_="tit")
```

```python
info= soup.find("div", class_="article").find_all("dt",class_="tit")
info[0].find("a").text
```

```python
info= soup.find("div", class_="article").find_all("dt",class_="tit")
for i in info:
    title=i.find("a").text
    print(title)
```
