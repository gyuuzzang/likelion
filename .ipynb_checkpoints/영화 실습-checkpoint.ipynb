```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
basic_url= " https://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword=187322"

url1=" https://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword=187322&page=1"
```

```python
page = urlopen(url1)
soup = BeautifulSoup(page, "html.parser")
comment_all = soup.find_all('td', class_='title')
```

```python
import time

comments = []
cnt = 0

for i in range(1,50):
    url = basic_url + '&page=' + str(i)
    page = urlopen(url)
    soup = BeautifulSoup(page,'html.parser')
    
    comment_all = soup.find_all('td', class_='title')
    
    for comment in comment_all:
        temp= list(comment.children)[6].strip()
        comments.append(temp)
        
    time.sleep(1)
    
print(len(comments))
```

```python
import pandas as pd
import os
```

```python
dict_doc={"text":comments}
doc = pd.DataFrame(dict_doc)
doc.to_csv("크루엘라리뷰.csv",index=False)

os.listdir(os.getcwd())
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = "https://movie.naver.com/movie/running/current.nhn"
page = urlopen(url)
soup = BeautifulSoup(page, 'lxml')
```

```python
name1 =soup.find_all("dt",class_="tit")
name1[0].find("a").text
```

```python
name1 =soup.find_all("dt",class_="tit")
for one in name1:
    name= one.find("a").text
    print(name)
```

```python
movie_title = []
for one in name1:
    name= one.find("a").text
    movie_title.append(name)
```

```python
rate1 =soup.find_all("dd",class_="star")
for one in rate1:
    rate= one.find_all("span", class_="num")[1].text
    print(rate)
```

```python
people =soup.find("dd",class_="star").find("span", class_="num2").text
people
```

```python
rate =soup.find_all("dd",class_="star")[0].find_all("span", class_="num")[0].text
people =soup.find_all("dd",class_="star")[0].find_all("span", class_="num")[1].text
print("평점:",rate, "예매율",people,"%")
```

```python
rate1 =soup.find_all("dd",class_="star")
for a in rate1:
    rate=a.find_all("span", class_="num")[0].text
    people=a.find_all("span", class_="num")[1].text
    par=a.find("span", class_="num2").text
    print("평점:",rate, "예매율",people,"%",par)
```
